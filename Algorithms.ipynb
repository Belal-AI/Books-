```python
def binarysearch(listof,item):
    low=0
    high=len(listof)-1
    
    while(low<=high):
        mid=int((high+low)/2)
        print(mid)
        if(listof[mid]==item):
            return item
        elif(listof[mid]>item):
            high=mid-1
        else:
            low=mid+1
    return 'sezar'
    
    
```



*   **Selection Sort**

*   ** get the smallest or largest element and append it on array (n^2)** 




```python
def getsmallest(lista):
    min_lista=lista[0]
    index=0
    for i in range(1,len(lista)):
        if lista[i]<min_lista:
            min_lista=lista[i]
            index=i
    return index
        
```

```python
def selec_sort(lista):
    arrsorted=[]
    for i in range(len(lista)):
        min_indx=getsmallest(lista)
        arrsorted.append(lista.pop(min_indx))     ##remove the min value of the array 
                                                 ##and pass the array again without the min 
                                               ##before to get the new min and so on 
    return arrsorted
        
        
    
```

```python
arr=[3,4,55,3,22,33,3,113,22]
```

```python
selec_sort(arr)
```

**Recursion**

```python
def factorial(X):
    if X==1:
        return X
    else:
        return X*factorial(X-1)
        
```

```python
factorial(5)
```

**Quicksort algo** 



*  *Quicksort is unique because its speed depends on the pivot you choose. 
Before I talk about quicksort, lets look at the most common Big O run 
times again.*


*  Theres another sorting algorithm called merge sort, which is 
O(n log n). Much faster! Quicksort is a tricky case. In the worst case, 
quicksort takes O(n2) time.
Its as slow as selection sort! But thats the worst case. In the average 
case, quicksort takes O(n log n) time. So you might be wondering:





```python
def quicksort(lista):
  if (len(lista)<2):
    return lista
  else:
    bivot=lista[0]
    greater=[i for i in  lista[1:] if i>bivot]
    smaller=[i for i in  lista[1:] if i<bivot]
    return quicksort(smaller)+[bivot]+quicksort(greater)
 







```

```python
quicksort([30,10,-1,2,3])
```
